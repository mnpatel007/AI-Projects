```python
#simple linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Define independent and dependent variables
X = df[['tenure']]
y = df['MonthlyCharges']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# Encode the 'Churn' column to binary (1 for Yes, 0 for No)
df['Churn'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

# Independent and dependent variables
X_simple = df[['MonthlyCharges']]
y_simple = df['Churn']

# Split the data into train (65%) and test (35%)
X_train, X_test, y_train, y_test = train_test_split(X_simple, y_simple, test_size=0.35, random_state=42)

# Build and train the logistic regression model
log_model_simple = LogisticRegression()
log_model_simple.fit(X_train, y_train)

# Predict values on the test set
y_pred_simple = log_model_simple.predict(X_test)

# Build confusion matrix and calculate accuracy score
conf_matrix_simple = confusion_matrix(y_test, y_pred_simple)
accuracy_simple = accuracy_score(y_test, y_pred_simple)

# Display results
print("Confusion Matrix (Simple Logistic Regression):\n", conf_matrix_simple)
print(f"Accuracy (Simple Logistic Regression): {accuracy_simple * 100:.2f}%")
```

```python
#multiple logistic regression
# Independent and dependent variables
X_multi = df[['tenure', 'MonthlyCharges']]
y_multi = df['Churn']

# Split the data into train (80%) and test (20%)
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, test_size=0.2, random_state=42)

# Build and train the logistic regression model
log_model_multi = LogisticRegression()
log_model_multi.fit(X_train_multi, y_train_multi)

# Predict values on the test set
y_pred_multi = log_model_multi.predict(X_test_multi)

# Build confusion matrix and calculate accuracy score
conf_matrix_multi = confusion_matrix(y_test_multi, y_pred_multi)
accuracy_multi = accuracy_score(y_test_multi, y_pred_multi)

# Display results
print("Confusion Matrix (Multiple Logistic Regression):\n", conf_matrix_multi)
print(f"Accuracy (Multiple Logistic Regression): {accuracy_multi * 100:.2f}%")
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

# Independent and dependent variables
X_tree = df[['tenure']]  # Independent variable
y_tree = df['Churn']     # Dependent variable (Churn already encoded as 1/0)

# Split the dataset into 80% training and 20% testing
X_train_tree, X_test_tree, y_train_tree, y_test_tree = train_test_split(X_tree, y_tree, test_size=0.2, random_state=42)

# Build and train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_tree, y_train_tree)

# Predict values on the test set
y_pred_tree = dt_model.predict(X_test_tree)

# Confusion matrix and accuracy score
conf_matrix_tree = confusion_matrix(y_test_tree, y_pred_tree)
accuracy_tree = accuracy_score(y_test_tree, y_pred_tree)

# Display results
print("Confusion Matrix (Decision Tree):\n", conf_matrix_tree)
print(f"Accuracy (Decision Tree): {accuracy_tree * 100:.2f}%")
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

# Independent and dependent variables
X_rf = df[['tenure', 'MonthlyCharges']]  # Independent variables
y_rf = df['Churn']                       # Dependent variable (Churn already encoded as 1/0)

# Split the dataset into 70% training and 30% testing
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(X_rf, y_rf, test_size=0.3, random_state=42)

# Build and train the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_rf, y_train_rf)

# Predict values on the test set
y_pred_rf = rf_model.predict(X_test_rf)

# Confusion matrix and accuracy score
conf_matrix_rf = confusion_matrix(y_test_rf, y_pred_rf)
accuracy_rf = accuracy_score(y_test_rf, y_pred_rf)

# Display results
print("Confusion Matrix (Random Forest):\n", conf_matrix_rf)
print(f"Accuracy (Random Forest): {accuracy_rf * 100:.2f}%")
```
